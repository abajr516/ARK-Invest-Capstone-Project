````python
# cleanup() - removes all symbols and letters from the columns so that they can be converted to floats
def cleanup(s):
  for i in s:
    if (i==','):
      s = s.replace(i, '')
    if (i=='%'):
      s = s.replace(i, '')
      s=float(s)/100
    if (i=='B'):
      s = s.replace(i, '')
      s=float(s) * 1000000000
    if (i=='M'):
      s = s.replace(i, '')
      s=float(s) * 1000000
    if (i=='K'):
      s = s.replace(i, '')
      s=float(s) * 1000
  return float(s)

# convertStrToFloat() - calls cleanup() and handles columns that have two values that need to be averaged or differenced
def convertStrToFloat(s):
  if (s=='-'):
    return np.nan
  elif ' - ' in s:
    lst = s.split(' - ')
    lst1 = []
    for p in lst:
      p = cleanup(p)
      lst1.append(p)
    s = lst1[0] - lst1[1]
    return float(s)
  elif ' ' in s:
    lst2 = s.split(' ')
    lst3 = []
    for p in lst2:
      p = cleanup(p)
      lst3.append(p)
    s = (lst3[0] + lst3[1])/2
    return float(s)
  else:
    return cleanup(s)
````

## Optional Branch: TF-IDF (Another Unsupervised Association)

Reference: https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfVectorizer.html

Reference: https://www.capitalone.com/tech/machine-learning/understanding-tf-idf/

TF-IDF can be broken down into two parts TF (term frequency) and IDF (inverse document frequency).

## TF (term frequency)
  * Proportion a word or term appears in a document.

## IDF (inverse document frequency)
  * Proportion a word or term appears in a corpus.
  * A corpus just means the universe of all documents.
  
## Recall, Bayesian vs Frequentist
  * Frequentist - if something has never happened, it will never happen.
  * Bayesian - if something has never happened, we have to adjust our prior belief to an extremely low probability.

Cons of using kNN:
kNN has high variance because it relies on finding neighbors which means there is typically a very small number of points to be in the average. Law of large numbers does not apply to small samples, and problem gets exponentially worse when p is large. When dimension p is large (i.e. many features), the distance to the nearest neighbor increases exponentially, given the same number of data points.

### Linear Regression Helper

* Calculate alphas and betas on a univariate regression on all columns of a DataFrame.
* Example to evaluate new ARKK using alpha/beta to the original

````python
import statsmodels.api as sm
def calcOLSDf(calcrets,indp):
  results={}
  for c in calcrets.columns:
    X2 = sm.add_constant(calcrets[indp])
    mod = sm.OLS(calcrets[c],X2)
    res = mod.fit()
    results[c]=res.params
  pd.options.display.float_format = '{:,.4f}'.format
  return pd.DataFrame(results).T.sort_values(indp)
````

````python
alphabetaToArk=calcOLSDf(calcrets,'ARKK')
````
